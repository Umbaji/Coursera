```python
# import required module
import os
import pandas as pd
import shutil

# assign directory
directory = "D:\Downloads\Coursera_Deep_Learning_Cert"

# iterate over files in
# that directory
class github_upload():
    def __init__(self, directory):
        self.dir_name = directory
        self.locations = []
        
    def locate_ftype(self):
        content = []
        locations = []
        temp_dict = dict()
        folder =""
        for root,dirs,files in os.walk(directory):
            for dir in dirs:
                #print(os.path.join(root,dir))
                folder = str(os.path.join(root,dir))
                content.append(folder)
                
        for element in content:
            for root,dirs,files in os.walk(element):
                notebooks =[]
                for file in files :
                    if "ipynb" in str(file):
                        notebooks.append(file)
                if (len(notebooks)!=0):
                    temp_dict = {element : notebooks}
                    #print(temp_dict)
                    locations.append(temp_dict)

        self.locations = locations
                
        return locations

    def folder_to_push(self):
        push_loc = self.dir_name+"\\push_folder"
        try : 
            os.makedirs(push_loc)
        except OSError:
            print ("Creation of the directory %s failed" %push_loc)
        else:
            print ("Successfully created the directory %s " %push_loc)
        
        #sub_dir = os.path.join(root,"sub_dir")
        os.chdir(push_loc)
        for element in self.locations:
            for key in element :
                print(key)
                sub_dir_to_cp = key
                dir_elem= sub_dir_to_cp.lstrip("D:\Downloads\Coursera_Deep_Learning_Cert")
                new_sub_dir = push_loc +"\\"+dir_elem
                print(new_sub_dir)
                try : 
                    os.makedirs(new_sub_dir) 
                except : 
                    print("%s already created",new_sub_dir)
                os.chdir(new_sub_dir)
                values_to_cp = element[key]
                print(values_to_cp)
                for elem in values_to_cp:
                    src = key + "\\" + elem                           
                    print('File location :',src)
                    dest_file = new_sub_dir + "\\"+ elem
                    print('File destination :',dest_file)
                    try : 
                        shutil.copyfile(src,dest_file)
                        print ("Copy Successful")
                    except OSError:
                        print ("Copy failed")
                    
        
        return True
    
    
if __name__=="__main__" : 
    
    upload_in = github_upload(directory)
    locations = upload_in.locate_ftype()
    print(locations)
    upload_in.folder_to_push()
                        
                    
            
            
        
        
#for filename in os.listdir(directory):
#	f = os.path.join(directory, filename)
#	# checking if it is a file
#	if os.path.isfile(f):
#		print(f)

# iterate over files in
# that directory
#for root, dirs, files in os.walk(directory):
        

        

    
        

```
